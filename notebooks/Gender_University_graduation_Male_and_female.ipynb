## Question 1: Do women have a larger mean of degrees than men?

```python
# Select the top 10 universities with the highest number of awards for men
top_men = degree_df.nlargest(10, 'awards_men').copy()

# Select the least 10 universities with the lowest number of awards for men
least_men = degree_df.nsmallest(10, 'awards_men').copy()

# Select the top 10 universities with the highest number of awards for women
top_women = degree_df.nlargest(10, 'awards_women').copy()

# Select the least 10 universities with the lowest number of awards for women
least_women = degree_df.nsmallest(10, 'awards_women').copy()

# Add category labels
top_men['category'] = 'Top 10 Male'
least_men['category'] = 'Least 10 Male'
top_women['category'] = 'Top 10 Female'
least_women['category'] = 'Least 10 Female'
```

```python
# Assuming 'top_men', 'least_men', 'top_women', 'least_women' already exist
combined_table = pd.concat([top_men, least_men, top_women, least_women], ignore_index=True)

# Merge with university data to get state abbreviations
combined_table = pd.merge(combined_table, uni_df[['university_name', 'STABBR']], on='university_name', how='left')

# Rename 'STABBR' to 'state'
combined_table.rename(columns={'STABBR': 'state'}, inplace=True)

# Reorder columns for clarity
combined_table = combined_table[['category', 'university_name', 'state', 'awards_men', 'awards_women']]

# Fill NaN values with 0
combined_table.fillna(0, inplace=True)
```

```python
# Assuming 'combined_table' is the DataFrame you are working with
# Calculate the mean number of awards for men and women
mean_awards = combined_table[['awards_men', 'awards_women']].mean()

# Print the mean awards
print("Mean Number of Awards for Men and Women:")
print(mean_awards)

# Create a bar graph for mean awards for men and women
mean_awards.plot(kind='bar', color=['#74A8D4', '#EC6A5A'])

# Add labels and title
plt.xlabel('Men                                     Women', fontsize=12)
plt.ylabel('Average Degrees Earned', fontsize=12)

# Remove x-axis labels (awards_men and awards_women)
ax = plt.gca()  # Get current axes
ax.set_xticklabels([])  # Remove x-axis tick labels

# Make the x-axis labels horizontal
plt.xticks(rotation=0)  # Horizontal x-axis labels

# Remove the grid
plt.grid(False)

# Set background color
fig = plt.gcf()  # Get the current figure
fig.patch.set_facecolor('#FEFCF5')  # Background color for figure
ax.set_facecolor('#FEFCF5')         # Background color for axes

# Show the plot
plt.show()
```

## Question 6: Do institution size, university type, and HBCU have influence on university choice based on gender?

```python
# Ensure 'INSTSIZE' is renamed to 'institution_size' in uni_df
uni_df.rename(columns={'INSTSIZE': 'institution_size'}, inplace=True)

# Merge uni_df and degree_df on 'institution_id' to combine institution size with awards data
merged_df = pd.merge(degree_df, uni_df[['institution_id', 'institution_size']], on='institution_id', how='inner')

# Group by institution size and calculate mean awards for males and females
institution_awards = merged_df.groupby('institution_size').agg({
    'awards_men': 'mean',
    'awards_women': 'mean'
}).reset_index()

# Melt the dataframe to a long format for combined visualization
institution_awards_long = pd.melt(
    institution_awards,
    id_vars=['institution_size'],
    value_vars=['awards_men', 'awards_women'],
    var_name='gender',
    value_name='mean_awards'
)

# Map gender column to more readable labels
institution_awards_long['gender'] = institution_awards_long['gender'].replace({
    'awards_men': 'Male',
    'awards_women': 'Female'
})

# Combined bar graph for male and female awards
fig_combined = px.bar(
    institution_awards_long,
    x='institution_size',
    y='mean_awards',
    color='gender',
    barmode='group',
    title='Mean Awards by Institution Size and Gender',
    labels={'institution_size': 'Institution Size', 'mean_awards': 'Mean Number of Awards', 'gender': 'Gender'},
    color_discrete_sequence=['#74A8D4', '#EC6A5A']
)

fig_combined.update_layout(
    xaxis_tickangle=-45,
    xaxis_title='Institution Size',
    yaxis_title='Mean Number of Awards',
)

# Show the graph
fig_combined.show()
```

```python
# Rename 'HBCU' to 'hbcu' in uni_df
uni_df.rename(columns={'HBCU': 'hbcu'}, inplace=True)

# Merge uni_df and degree_df on 'institution_id' to combine HBCU status with awards data
merged_df = pd.merge(degree_df, uni_df[['institution_id', 'hbcu']], on='institution_id', how='inner')

# Group by HBCU status and calculate mean awards for males and females
hbcu_awards = merged_df.groupby('hbcu').agg({
    'awards_men': 'mean',
    'awards_women': 'mean'
}).reset_index()

# Map HBCU values to readable labels (e.g., 1 -> "HBCU", 0 -> "Non-HBCU")
hbcu_awards['hbcu'] = hbcu_awards['hbcu'].replace({1: 'HBCU', 2: 'Non-HBCU'})

# Melt the dataframe to long format for combined visualization
hbcu_awards_long = pd.melt(
    hbcu_awards,
    id_vars=['hbcu'],
    value_vars=['awards_men', 'awards_women'],
    var_name='gender',
    value_name='mean_awards'
)

# Map gender column to readable labels
hbcu_awards_long['gender'] = hbcu_awards_long['gender'].replace({
    'awards_men': 'Male',
    'awards_women': 'Female'
})

# Combined bar graph for male and female awards by HBCU status
fig_hbcu = px.bar(
    hbcu_awards_long,
    x='hbcu',
    y='mean_awards',
    color='gender',
    barmode='group',
    title='Mean Awards by HBCU Status and Gender',
    labels={'hbcu': 'HBCU Status', 'mean_awards': 'Mean Number of Awards', 'gender': 'Gender'},
    color_discrete_sequence=['#74A8D4', '#EC6A5A']
)

fig_hbcu.update_layout(
    xaxis_title='HBCU Status',
    yaxis_title='Mean Number of Awards',
)

# Show the graph
fig_hbcu.show()
```
